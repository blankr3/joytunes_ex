```python
import os
import glob
import numpy as np
import scipy.signal
import scipy.io.wavfile
from tqdm.notebook import tqdm
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Inputs
ref_path = r"D:\dev\joytunes\ref_samples.wav"
segment_paths = glob.glob(os.path.join("D:\dev\joytunes\segments", "*.wav"))
```

```python
# Load data
dataset = dict()
dataset['ref'] = dict()
dataset['ref']['sample_rate'], dataset['ref']['amps'] = scipy.io.wavfile.read(ref_path)  # load ref data
for p in tqdm(segment_paths):  # load segments data
    name = os.path.basename(p).split('.wav')[0]
    dataset[name] = dict()
    dataset[name]['sample_rate'], dataset[name]['amps'] = scipy.io.wavfile.read(p)
```

```python
# plot data
fig, axs = plt.subplots(ncols=1, nrows=len(dataset), figsize=(10,10))

for i, (name, data) in enumerate(dataset.items()):
    axs[i].plot(data['amps'], label=name)
    axs[i].set_xlabel('samples')
    axs[i].set_ylabel('amplitude')
    axs[i].legend()

fig.tight_layout()
```

```python
# cross correlate

for name in dataset:
    res = scipy.signal.fftconvolve(dataset[name]['amps'][::-1], dataset['ref']['amps'], mode='full')  # convolve ref and segment arrays (flip the ref array)
    res /= np.max(res)  # normalize for convenience
    max_index = np.argmax(res)  # index of max correlation point
    max_correlation_sample = range(-len(res) // 2 , len(res) // 2)[max_index]

    # plot data
    fig, axs = plt.subplots(ncols=1, nrows=2, figsize=(10,6))
    axs[0].plot(res)
    axs[0].plot([max_correlation_sample, max_correlation_sample], [-1.0, 1.0])
    axs[0].set_title('Cross Correlation with ref')
    
    shifted_x = list(range(max_correlation_sample, len(dataset[name]['amps']) + max_correlation_sample))
    axs[1].plot(shifted_x, dataset['ref']['amps'][shifted_x])
    axs[1].plot(shifted_x, dataset[name]['amps'])
    axs[1].set_title('Overlap segment on ref')
    axs[1].set_xlabel('samples')
    fig.tight_layout()
```
